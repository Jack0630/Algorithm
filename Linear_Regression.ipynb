```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import load_boston
```

```python
boston=load_boston()
```

```python
boston.feature_names
```

```python
x=boston.data
y=boston.target
```

```python
len(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x, y , test_size=0.3, random_state=87)
```

```python
regr=LinearRegression()
```

```python
regr.fit(x_train,y_train)
```

```python
y_predict=regr.predict(x_test)
```

```python
plt.scatter(y_test,y_predict)
plt.plot([0,50],[0,50],'r')
plt.xlabel('True Price')
plt.ylabel('Predicted Price')
plt.show()
```

左上是模型預測的價格(regr.predict(x_test))，右下是實際的價格(y_test)，可以看出當價格在40元以下，模型和實際價格並沒有太大差異，但是當價格超過40元，預測價格會偏低

```python
plt.figure(figsize=(8,10))
for k , feature in enumerate(boston.feature_names):
    plt.subplot(5,3,k+1)
    plt.scatter(x[:,k], y, s=3)
    plt.ylabel('price')
    plt.xlabel(feature)
    plt.tight_layout()
```

[小技巧]一次畫多個圖

```python
x= np.linspace(0,10,20)
```

```python
#2*2 第一張
plt.subplot(2,2,1)
plt.plot(x,np.sin(x))

#2*2 第二張
plt.subplot(2,2,2)
plt.plot(x,np.cos(x))

#2*2 第三張
plt.subplot(2,2,3)
plt.plot(x,x**2)

#2*2 第四張
plt.subplot(2,2,4)
plt.plot(x,x+x**2-1)
```
